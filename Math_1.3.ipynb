3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности:

```python
# уравнение окружности: (x - x0)^2 + (y - y0)^2 = R^2
# отсюда y = (R^2 - (x - x0)^2)^0.5 + y0
R = 2
x0 = -1
y0 = 1
x = np.linspace(-R + x0, R + x0, 100)
y = np.sqrt(R**2 - (x - x0)**2) + y0
y1 = -np.sqrt(R**2 - (x - x0)**2) + y0
plt.plot(x, y)
plt.plot(x, y1)
plt.axis('scaled')
plt.ylim((-R + y0)*1.1, (R + y0)*1.1)
```

эллипса:

```python
# уравнение элипса: ((x - x0)^2)/a + ((y - y0)^2)/b = 1
# отсюда y = (1 - b/a*(x - x0)^2)^0.5 + y0
a = 3
b = 2
x0 = -1
y0 = 1
x = np.linspace(-np.sqrt(a/b) + x0, np.sqrt(a/b) + x0, 100)
y = np.sqrt(1 - b/a*(x - x0)**2) + y0
y1 = -np.sqrt(1 - b/a*(x - x0)**2) + y0
plt.plot(x, y)
plt.plot(x, y1)
plt.axis('scaled')
plt.ylim((-np.sqrt(b) + y0)*1.1, (np.sqrt(b) + y0)*1.1)
```

```python
гиперболы:
```

```python
# уравнение гиперболы: ((x - x0)^2)/a - ((y - y0)^2)/b = 1
# отсюда y = (1 + b/a*(x - x0)^2)^0.5 + y0
a = 3
b = 2
x0 = -1
y0 = 1
x = np.linspace(-np.sqrt(a/b) + x0, np.sqrt(a/b) + x0, 100)
y = np.sqrt(1 + b/a*(x - x0)**2) + y0
y1 = -np.sqrt(1 + b/a*(x - x0)**2) + y0
plt.plot(x, y)
plt.plot(x, y1)
plt.axis('scaled')
plt.ylim((-np.sqrt(b) + y0)*1.1, (np.sqrt(b) + y0)*1.1)
```

4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

```python
# подставим значение y во второе уравнение: exp(x) + x*(1 - x^2 + 1) = 1 -> exp(x)- x^3 + 2*x - 1 = 0
# посроим график данного уравнения:
def f(x):
    return np.exp(x) - x**3 +2*x - 1

n = 100
x0, xn = -3, 5
x = np.linspace(x0, xn, n)
y = f(x)
plt.hlines(0, x0, xn)
plt.vlines(0, min(y), max(y))
plt.plot(x, y)
```

```python
from scipy.optimize import fsolve

x_solv1 = fsolve(f, -2)
x_solv2 = fsolve(f, -0.5)
x_solv3 = fsolve(f, 2)
x_solv4 = fsolve(f, 4)
print(x_solv1, x_solv2, x_solv3, x_solv4)
```

Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1

```python
# в предыдущем задании мы нашли корни уравнения. Теперь найдем значение функции на каждом из отрезков:
x_solv = [x_solv1[0], x_solv2[0], x_solv3[0], x_solv4[0]]
print("Решение системы неравенств выполняется на отрезках")
if f(x_solv[0] - 1) > 0:
    print(f"(-inf, {x_solv[0]})")
for i in range(len(x_solv) - 1):
    if f((x_solv[i+1] + x_solv[i])/2) > 0:
        print(f"({x_solv[i]}, {x_solv[i+1]})")
if f(x_solv[len(x_solv)-1] + 1) > 0:
    print(f"({x_solv[0]}, +inf)")
```
